```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

# 한글 폰트 경로 지정
font_path = '/Users/mac/Desktop/빅데이터 프로젝트/project data/KBIZ한마음명조 B.otf'
font_prop = font_manager.FontProperties(fname=font_path)


# 연도 리스트 (2017년부터 2021년)
years = list(range(2017, 2022))

# 새로운 리스트 생성 및 매핑
sev_reversed = seoul_sev[:5][::-1]

# 그래프 그리기
plt.figure(figsize=(16, 9))
line, = plt.plot(years, sev_reversed, marker='o', linestyle='-', color='#4CAF50', label='중증 환자 수')  # 색상 추가

# 그래프에 미적 요소 추가
plt.title('중증 환자 추이\n- 서울 -\n ', fontproperties=font_prop, fontsize=25, fontweight='bold')  # 제목에 더 크게, 굵게, 색상 변경
plt.xlabel('연도', fontproperties=font_prop, fontsize=18, color='#666666')  # x축 라벨에 색상 변경
plt.ylabel('중증 환자 수', fontproperties=font_prop, fontsize=18, color='#666666')  # y축 라벨에 색상 변경
plt.xticks(years, ['2017년', '2018년', '2019년', '2020년', '2021년'], fontproperties=font_prop, fontsize=15, color='#666666')  # x축 눈금에 색상 변경

# 그리드 추가
plt.grid(True, linestyle='--', alpha=0.5, color='#AAAAAA')

# 주석 추가
for i, txt in enumerate(sev_reversed):
    plt.annotate(f'{int(txt):,}', (years[i], sev_reversed[i]), textcoords="offset points", xytext=(0, 10), ha='center', color='#333333')  # 주석에 색상 변경

# 범례 추가
plt.legend(prop=font_prop, loc='upper left', fontsize=12, frameon=False)

# 배경 색상 변경
plt.gca().set_facecolor('#F5F5F5')

# 라벨 텍스트에 폰트 적용
plt.gca().xaxis.set_tick_params(labelsize=12, labelcolor='#666666')
plt.gca().yaxis.set_tick_params(labelsize=12, labelcolor='#666666')

# y축 레이블 텍스트 일부 생략
ylabels = [f'{int(label/1000)}K' for label in plt.gca().get_yticks()]  # 예시: 186686 -> '186.7K'
plt.gca().set_yticklabels(ylabels, fontproperties=font_prop, fontsize=12, color='#666666')

# 중증 환자 수와 연도 텍스트 색상 변경
plt.gca().title.set_color('#333333')  # 제목
plt.gca().xaxis.label.set_color('#333333')  # x축 라벨
plt.gca().yaxis.label.set_color('#333333')  # y축 라벨
plt.gca().tick_params(axis='both', colors='#666666')  # 눈금 색상 변경

plt.show()
```
