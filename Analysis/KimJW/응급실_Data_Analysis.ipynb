```python
import matplotlib.pyplot as plt
import numpy as np

def plot_population_by_age(dataframe):
    # 원하는 범위의 칼럼 선택
    age_columns = dataframe.loc[:, '0-9세':'100세 이상']

    # 그래프 크기 설정 (크게 조절)
    plt.figure(figsize=(16, 12))

    # 연령별로 다른 색상 지정
    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'brown', 'pink', 'gray', 'cyan']

    # 도시 이름을 x축으로, 연령대를 색상으로 표현하는 수평 막대 그래프 플로팅
    for i, city in enumerate(dataframe['도시이름']):
        # 막대 그래프의 시작 위치를 나타내는 변수
        left = 0
        print(i)
        # 각 연령대에 대해 색상을 지정하여 막대 그래프 플로팅
        for j, age_group in enumerate(age_columns.columns):
            
            plt.barh(city, age_columns[age_group][i], left=left, color=colors[j])
            left += age_columns[age_group][i]

    # 그래프 제목 및 축 레이블 설정
    plt.title('2021 년도 도시별 연령대별 인구비')
    plt.xlabel('인구수')
    plt.ylabel('도시')

    # 색상 범례 추가
    legend_labels = age_columns.columns
    plt.legend(legend_labels, loc='upper left', bbox_to_anchor=(1, 1), title='연령대')

    # 그리드 표시
    plt.grid(True)

    # 그래프 표시
    plt.tight_layout()  # 그래프 크기 조절 후 레이아웃 최적화
    plt.show()

# 함수 호출
plot_population_by_age(selected_columns)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_population_by_age(dataframe):
    # 원하는 범위의 칼럼 선택
    age_columns = dataframe.loc[:, '0-9세':'100세 이상']

    # 그래프 크기 설정 (크게 조절)
    plt.figure(figsize=(16, 12))

    # 도시 이름을 x축으로, 연령대를 색상으로 표현하는 수평 막대 그래프 플로팅
    for i, city in enumerate(dataframe['도시이름']):
        # 다양한 색상 사용
        colors = plt.cm.viridis(np.linspace(0.0, 1.0, len(age_columns.columns)))
        plt.barh(city, age_columns.iloc[i], color=colors)

    # 그래프 제목 및 축 레이블 설정
    plt.title('2021 년도 도시별 연령대별 인구비')
    plt.xlabel('인구수')
    plt.ylabel('도시')

    # 색상 범례 추가
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='연령대', labels=age_columns.columns)

    # 그리드 표시
    plt.grid(True)

    # 그래프 표시
    plt.tight_layout()  # 그래프 크기 조절 후 레이아웃 최적화
    plt.show()

# 함수 호출
plot_population_by_age(selected_columns)
```
