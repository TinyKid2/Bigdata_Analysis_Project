```python
import numpy as np  # Import NumPy for handling numerical operations
import pandas as pd  # Import Pandas for data manipulation and analysis
import warnings  # Import Warnings to suppress unnecessary warnings

# Suppress warning messages
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn
from matplotlib import rc
import os
rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel("/Users/stone/Desktop/필요/4_2/BigData/팀플/데이터/12.7/응급의료기관+병상+현황+-+시도별,+연도별.csv")
```

```python
df_1 = pd.DataFrame()
```

```python
df.head()
```

```python
years = ['2017년.1', '2018년.1', '2019년.1', '2020년.1', '2021년.1']
for i in years:
    df_1[i] = df[i]
df_1['분류'] = df['분류']
```

```python
cities = [
    '서울', '부산', '대구', '인천', '광주', '대전', '울산',
    '세종', '경기', '강원', '충북', '충남', '전북', '전남',
    '경북', '경남', '제주'
]
year = ['2017년', '2018년', '2019년', '2020년', '2021년']
```

```python
df_1.head(30)
```

```python
df_1.fillna(0, inplace=True)
df_1.replace(',', '', regex=True, inplace=True)
```

```python
# df_City_E_have_1_new 데이터프레임에서 '서울' 행 선택
seoul_data = df_1.loc[5+0]
seoul_data.drop('분류',inplace=True)
# 시각화
plt.figure(figsize=(6, 4))

# '서울' 행에 대한 선 그래프 그리기
plt.plot(year, seoul_data, marker='o', label=f'{cities[0]}')

plt.title(f'{cities[0]} 응급의료기관 병상 추이')
plt.xlabel('년도')
plt.ylabel('병상수')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
